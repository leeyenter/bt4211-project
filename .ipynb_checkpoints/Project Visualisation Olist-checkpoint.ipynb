```python
import pandas as pd
import numpy as np
import pandas_profiling as pp
import os

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Read in all datasets

```python
files = os.listdir('./brazilian-ecommerce/')

res = dict(zip([x.split('.')[0] for x in files],
               [pd.read_csv('./brazilian-ecommerce/' + x) for x in files]))

files
```

```python
mql = res['olist_marketing_qualified_leads_dataset']
mql.info()
```

```python
deals = res['olist_closed_deals_dataset']
deals.info()
```

```python
sellers = res['olist_sellers_dataset']
sellers.info()
```

# Exploring seller specific data

## qualified leads dataset

* mql_id - Marketing Qualified Lead id
* first_contact_date - Date of the first contact solicitation.
* landing_page_id - Landing page id where the lead was acquired
* origin - Type of media where the lead was acquired

```python
mql['first_contact_date'] = mql['first_contact_date'].apply(pd.to_datetime)
mql.head()
```

```python
# top channels that provide leads
mql['origin'].value_counts().plot.barh(figsize=(8,5));
```

```python
# most came in 2018
mql['first_contact_date'].dt.year.value_counts()
```

```python
mql['first_contact_date'].dt.dayofweek.value_counts() # Monday=0, Sunday=6
```

```python
# count of traffic through the week split by traffic origin
mql['day_of_contact'] = mql['first_contact_date'].dt.dayofweek
g = sns.FacetGrid(data = mql.groupby('origin')['day_of_contact'].value_counts()\
                            .reset_index(level='origin')\
                            .rename(columns={'day_of_contact':'count'})\
                            .reset_index(level='day_of_contact'),
                  col = 'origin', col_wrap = 3, aspect = 1.2, height = 3.5)
g = g.map(sns.barplot, 'day_of_contact', 'count')
```

```python
mql['landing_page_id'].value_counts().reset_index()[:30].plot()
```

The top 5-10 landing pages account for a very significant proportion of the leads. There is a very skewed long tail.

## closed deals dataset

* mql_id - Marketing Qualified Lead id
* seller_id - Seller id
* sdr_id - Sales Development Representative id
* sr_id - Sales Representative
* won_date - Date the deal was closed.
* business_segment - Lead business segment. Informed on contact.
* lead_type - Lead type. Informed on contact.
* lead_behaviour_profile - Lead behaviour profile. SDR identify it on contact.
* has_company - Does the lead have a company (formal documentation)?
* has_gtin - Does the lead have Global Trade Item Number (barcode) for his products?
* average_stock - Lead declared average stock. Informed on contact.
* business_type - Type of business (reseller/manufacturer etc.)
* declared_product_catalog_size - Lead declared catalog size. Informed on contact.
* declared_monthly_revenue - Lead declared estimated monthly revenue. Informed on contact.

```python
pp.ProfileReport(deals)
```

```python
deals['business_segment'].value_counts()[:30].reset_index().plot()
plt.xlabel('Top business segments')
plt.ylabel('Count');
```

```python
# merge with mql
marketing_funnel = mql.merge(deals, on='mql_id', how='left')
marketing_funnel.info() # only 10% get closed
```

```python
marketing_funnel.lead_behaviour_profile.value_counts()
```

```python
top_segm = marketing_funnel.business_segment.value_counts()[:6].index.tolist()
g = sns.FacetGrid(data = marketing_funnel\
                            .loc[marketing_funnel.business_segment.isin(top_segm)]\
                            .groupby('business_segment')['day_of_contact']\
                            .value_counts()\
                            .reset_index(level='business_segment')\
                            .rename(columns={'day_of_contact':'count'})\
                            .reset_index(level='day_of_contact'),
                  col = 'business_segment', col_wrap = 3, aspect = 1.2, height = 3.5)
g = g.map(sns.barplot, 'day_of_contact', 'count')
```

```python
# merge with marketing funnel
marketing_funnel = marketing_funnel.merge(sellers, on='seller_id', how='left')
marketing_funnel.info()
```

```python
# of those who didn't close, what is their profile
churned = marketing_funnel[marketing_funnel.won_date.isna()]
origin_churn = churned.origin.value_counts() / marketing_funnel.origin.value_counts()
origin_churn.sort_values(ascending=False)
```

```python
# out of the top landing pages, who contributed to churn the most?
landing_churn = churned.landing_page_id.value_counts()
landing_churn /= marketing_funnel.landing_page_id.value_counts()[landing_churn.index]
landing_churn = pd.DataFrame({'prop_churned':landing_churn, 
                              'total_landings':marketing_funnel.landing_page_id\
                                                .value_counts()[landing_churn.index]})
landing_churn.sort_values(ascending=False, by='total_landings')[:10]
```
